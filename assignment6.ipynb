Q1. Ask the user to enter a value (e.g., M) for a mathematical function. 
 Generate x values from -10 to 10 using np.linspace(). 
 Compute y values for: 
o y= M⋅x2 
o y=M⋅sin(x)

 Plot both functions in a single figure: 
o Use different colors and line styles. 
o Add legend, grid, and title.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

m=float(input("Enter the value for M:"))
x_values = np.linspace(-10,10,100)
y_values_1 = m * (x_values**2)
y_values_2 = m * np.sin(x_values)

plt.plot(x_values, y_values_1, 'r--', label=f'y = {m}*x^2')
plt.plot(x_values, y_values_2, 'b-', label=f'y = {m}*sin(x)')
plt.legend()
plt.grid()
plt.title("Mathematical Functions")
plt.show()
```

Q.2 Create a dataset of five subjects and scores. 
 Convert it into a Pandas DataFrame. 
 Plot the scores using a Seaborn bar plot with: 
o Different colors for each bar. 
o Annotations on top of each bar. 
o Title, axis labels, and grid. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
    'Score': [85, 92, 78, 88, 95]
}
df = pd.DataFrame(data)

ax = sns.barplot(data=df, x='Subject', y='Score', palette='viridis', hue='Subject')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), 
                textcoords='offset points')
    
plt.title('Scores by Subject', fontsize=16)
plt.xlabel('Subject', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and 
then generate a dataset of 50 values using np.random.randn(). 
 Create a 2x2 subplot layout displaying: 
o Line plot showing cumulative sum. 
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids. 

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 102316108
np.random.seed(roll_number)
data = np.random.randn(50)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Visualizations of Random Data', fontsize=16)

axs[0, 0].plot(np.cumsum(data), color='blue', label='Cumulative Sum')
axs[0, 0].set_title('Cumulative Sum of Data')
axs[0, 0].set_xlabel('Index')
axs[0, 0].set_ylabel('Cumulative Sum')
axs[0, 0].grid(True, linestyle='--', alpha=0.7)
axs[0, 0].legend()


noise = np.random.randn(50)
axs[0, 1].scatter(data, noise, c='r', alpha=0.7)
axs[0, 1].set_title("Scatter Plot with Random Noise")
axs[0, 1].set_xlabel("Generated Data")
axs[0, 1].set_ylabel("Random Noise")
axs[0, 1].grid(True)

plt.tight_layout()
plt.show()
```
