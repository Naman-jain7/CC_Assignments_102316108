```python
import pandas as pd
import numpy as np

roll_number = 102316108
np.random.seed(roll_number)

sales_data = np.random.randint(1000, 5001, size=(12, 4))

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
sales_df = pd.DataFrame(sales_data, columns=categories, index=months)
print(sales_df)
```

```python
print(sales_df.head())
print()
print(sales_df.describe())
print()
total_sales_per_category = sales_df.sum()
total_sales_per_month = sales_df.sum(axis=1)

sales_df['Total Sales'] = total_sales_per_month
sales_df['Growth Rate'] = sales_df['Total Sales'].pct_change() * 100

sales_df['Electronics'] *= 0.9

print()
print(sales_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
for category in categories:
    plt.plot(sales_df.index, sales_df[category], marker='o', label=category)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales Trends')
plt.legend()
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=sales_df[categories])
plt.title('Sales Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]]) 
i. Find element-wise absolute value 
ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each row. 
iii. Mean, Median and Standard Deviation of flattened array, of each column, and each row 

```python
import numpy as np
arr=np.array([[1, -2, 3],[-4, 5, -6]])

abs_array = np.abs(arr)
print(f"Element-wise absolute value: {abs_array}")
```

```python
flattened_percentiles = np.percentile(arr.flatten(), [25, 50, 75])
column_percentiles = np.percentile(arr, [25, 50, 75], axis=0)
row_percentiles = np.percentile(arr, [25, 50, 75], axis=1)
print(f"Flattened percentiles: {flattened_percentiles}")
print(f"Column percentiles: \n{column_percentiles}")
print(f"Row percentiles: \n{row_percentiles}")
```

```python
mean_flat = np.mean(arr)
median_flat = np.median(arr)
std_flat = np.std(arr)

mean_col = np.mean(arr, axis=0)
median_col = np.median(arr, axis=0)
std_col = np.std(arr, axis=0)

mean_row = np.mean(arr, axis=1)
median_row = np.median(arr, axis=1)
std_row = np.std(arr, axis=1)

print("\nMean (Flattened):", mean_flat)
print("Median (Flattened):", median_flat)
print("Std (Flattened):", std_flat)

print("\nMean (Columns):", mean_col)
print("Median (Columns):", median_col)
print("Std (Columns):", std_col)

print("\nMean (Rows):", mean_row)
print("Median (Rows):", median_row)
print("Std (Rows):", std_row)
```

Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).  
i. Find floor, ceiling and truncated value, rounded values 

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceil_values = np.ceil(a)
trunc_values = np.trunc(a)
rounded_values = np.round(a)

print("Original Array:", a)
print("Floor Values:   ", floor_values)
print("Ceil Values:    ", ceil_values)
print("Trunc Values:   ", trunc_values)
print("Rounded Values: ", rounded_values)
```

Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.

```python
def swap_elements(l, i, j):
    temp = l[i]
    l[i] = l[j]
    l[j] = temp

l = [1, 2, 3, 4, 5]
swap_elements(l, 1, 3)
print(l)
```

Q.5 Write a Python program to swap two elements in a set by conver ng it to a list first.

```python
def swap_set_elements(s, a, b):
    ls = list(s)
    if a in ls and b in ls:
        i, j = ls.index(a), ls.index(b)
        ls[i], ls[j] = ls[j], ls[i]
        return set(ls)
    return s

s = {10, 20, 30, 40, 50}
s = swap_set_elements(s, 20, 40)
print(s)
```
